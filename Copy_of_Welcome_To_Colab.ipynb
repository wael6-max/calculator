```
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Survived", y="Pclass", hue="Sex")
sns.regplot(data=df, x="Survived", y="Pclass", scatter=False, color='gray')
plt.title("Survived and Pclass")
plt.show()
```

```
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x="Age", y="Sex", hue="Pclass", marker="o")
plt.title("bigger gender")
plt.show()
```

```
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Age", y="Survived", palette="Set3")
plt.title("Ticket fare")
plt.show()
```

```
plt.figure(figsize=(8,5))
sns.barplot(data=df.groupby("Survived")["Pclass"].mean().reset_index(),
            x="Survived", y="Pclass", palette="pastel")
plt.title("titanic survival")
plt.show()
```

```
plt.figure(figsize=(8, 5))

plt.hist(df["Age"].dropna(), bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
```
